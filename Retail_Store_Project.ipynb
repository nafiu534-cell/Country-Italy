### Time Series Analysis on the Invoice Date 

```python
#Extracting some features in the InvoiceDate Column
Italy_info['InvoiceDay'] = Italy_info['InvoiceDate'].dt.date
Italy_info['InvoiceDayOfWeek'] = Italy_info['InvoiceDate'].dt.day_name()
Italy_info['InvoiceHour'] = Italy_info['InvoiceDate'].dt.hour
Italy_info['InvoiceMonth'] = Italy_info['InvoiceDate'].dt.month_name()
Italy_info['InvoiceYearMonth'] = Italy_info['InvoiceDate'].dt.to_period('M')

# Calculating total sales amount
Italy_info['TotalSales'] = Italy_info['Quantity'] * Italy_info['UnitPrice']

# Create day of week order for proper sorting
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
```

```python
# Total Sales Made by Month in Italy

plt.figure(figsize=(12, 6))
monthly_sales = Italy_info.groupby('InvoiceMonth')['TotalSales'].sum().reindex(month_order)
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='viridis')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Sales by Day of Week and by Hour of Day

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Daily pattern
daily_pattern = Italy_info.groupby('InvoiceDayOfWeek')['TotalSales'].sum().reindex(day_order)
sns.barplot(x=daily_pattern.index, y=daily_pattern.values, ax=ax1, palette='coolwarm')
ax1.set_title('Sales by Day of Week')
ax1.set_xlabel('Day of Week')
ax1.set_ylabel('Sales Amount')

# Hourly pattern
hourly_pattern = Italy_info.groupby('InvoiceHour')['TotalSales'].sum()
sns.barplot(x=hourly_pattern.index, y=hourly_pattern.values, ax=ax2, palette='coolwarm')
ax2.set_title('Sales by Hour of Day')
ax2.set_xlabel('Hour')
ax2.set_ylabel('Sales Amount')

plt.tight_layout()
plt.show()
```

```python
# Getting top 5 products by sales per month
top_products = Italy_info.groupby('Description')['TotalSales'].sum().nlargest(5).index

plt.figure(figsize=(14, 8))

# Plot monthly sales for top products
for product in top_products:
    product_sales = Italy_info[Italy_info['Description'] == product].groupby('InvoiceYearMonth')['TotalSales'].sum()
    plt.plot(product_sales.index.astype(str), product_sales.values, label=product, marker='o')

plt.title('Monthly Sales Trends for Top 5 Products')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Knowing the Transaction Status by Year Month

if 'Transaction Status' in Italy_info.columns:
    plt.figure(figsize=(14, 7))
    
    # Count transactions by status over time (monthly)
    status_counts = Italy_info.groupby(['InvoiceYearMonth', 'Transaction Status'])['InvoiceNo'].nunique().unstack()
    
    status_counts.plot(kind='bar', stacked=True, figsize=(14, 7))
    plt.title('Transaction Status Distribution Over Time')
    plt.xlabel('Month')
    plt.ylabel('Number of Transactions')
    plt.xticks(rotation=45)
    plt.legend(title='Transaction Status')
    plt.tight_layout()
    plt.show()
```

```python
# Get top 10 moving items
top_items = Italy_info.groupby('StockCode')['Quantity'].sum().nlargest(10).index

plt.figure(figsize=(14, 8))

# Plot monthly quantity for top items
for item in top_items:
    item_desc = Italy_info[Italy_info['StockCode'] == item]['Description'].iloc[0]
    item_qty = Italy_info[Italy_info['StockCode'] == item].groupby('InvoiceYearMonth')['Quantity'].sum()
    plt.plot(item_qty.index.astype(str), item_qty.values, label=f"{item} - {item_desc[:30]}...", marker='o')

plt.title('Monthly Quantity Sold for Top 10 Moving Items')
plt.xlabel('Month')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
